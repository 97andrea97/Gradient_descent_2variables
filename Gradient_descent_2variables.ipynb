```python
# Basic implementation of the gradient descent method to find the minimum of a 2-variables function.

import numpy as np

# define the functions and the derivatives:
f = lambda x1,x2: 2*(x1**2) + (x2**2) + x1*x2 - 6*x1 -  5*x2
df_x1 = lambda x1,x2: 4*x1 + x2 -6
df_x2 = lambda x1,x2: 2*x2 + x1 - 5
x0_1 = 1
x0_2 = 1
l_rate= 0.01

#gf = (df_x1(x0_1,x0_2), df_x2(x0_1,x0_2))


def gradient_2var(f,gf,l_rate,x0):
    stopping_rule = False
    x_old = x0
    count=0
    while not(stopping_rule):
        count+=1
        x_new = x_old - l_rate * gf
        #print("iteration n° {} critical point value: {}".format(count,x_new))
        if count>100:
            stopping_rule = True
        else:
            x_old=x_new
            gf = np.array([df_x1(x_old[0],x_old[1]), df_x2(x_old[0],x_old[1])])
            
    return x_new

critical_point= gradient_2var(f, np.array([df_x1(x0_1,x0_2), df_x2(x0_1,x0_2)]) , l_rate, (x0_1, x0_2))
print("the critical point computed with gradient descent is {}" .format(critical_point))   
    
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(-20, 20, 0.25)
Y = np.arange(-20, 20, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 2*(X**2) + (Y**2) + X*Y - 6*X - 5*Y

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,cmap='viridis', edgecolor='none')

ax.set_xlabel('X ')
ax.set_ylabel('Y ')
ax.set_zlabel('Z ')


x=critical_point[0]
y=critical_point[1]
z=2*(x**2) + (y**2) + x*y - 6*x - 5*y
ax.scatter(x, y, z, c="r", s=20)
plt.show()
```
